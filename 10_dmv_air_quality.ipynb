```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd

# Use raw string or double backslashes to avoid errors
data = pd.read_csv(r"D:\CL-1\DMV New\data.csv")

# Print the index of the DataFrame
print(data.index)
```

```python
print(data.isnull().sum())
print(data.shape)
data.info()
```

```python
print(data.isnull().sum())
data.tail()
```

```python
state_means = data.groupby('state')['no2'].mean()

x_axis = state_means.index
y_axis = state_means.values

plt.figure(figsize=(12, 6))
plt.bar(x_axis, y_axis, color='blue')
plt.xlabel('State')
plt.ylabel('Mean NO2 Value')
plt.title('Mean NO2 Values by State')
plt.xticks(rotation=90)  
plt.tight_layout()
plt.show()
```

```python
state_means = data.groupby('state')['so2'].mean()

x_axis = state_means.index
y_axis = state_means.values

plt.figure(figsize=(12, 6))
plt.bar(x_axis, y_axis, color='blue')
plt.xlabel('State')
plt.ylabel('Mean SO2 Value')
plt.title('Mean SO2 Values by State')
plt.xticks(rotation=90)  
plt.tight_layout()
plt.show()
```
