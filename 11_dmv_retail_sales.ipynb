```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv(r"D:\CL1\customer_shopping_data.csv")
df
```

```python
df.head()
```

```python
df.groupby("shopping_mall").count()
```

```python
df.groupby("category").count()
```

```python
branch_sales = df.groupby("shopping_mall").sum()
category_sales = df.groupby("category").sum()
branch_sales.sort_values(by = "price", ascending = False)
```

```python
category_sales.sort_values(by = "price", ascending = False)
```

```python
plt.figure(figsize=(6, 6))  # use plt.figure instead of plt.Figure
plt.pie(
    branch_sales["price"],
    labels=branch_sales.index,
    autopct="%1.1f%%",   # shows percentage with 1 decimal place
    startangle=90
)
plt.title("Sales Distribution by Branch")
plt.show()
```

```python
plt.figure(figsize=(6, 6))  # use plt.figure instead of plt.Figure
plt.pie(
    category_sales["price"],
    labels=category_sales.index,
    autopct="%1.1f%%",   # shows percentage with 1 decimal place
    startangle=90
)
plt.title("Sales Distribution by Category")
plt.show()
```

```python
mall_category_sales = df.groupby(['shopping_mall', 'category'])['price'].sum().unstack()

# Plot grouped bar chart
mall_category_sales.plot(kind='bar', figsize=(10, 6))
plt.title("Sales Amounts by Shopping Mall and Category")
plt.ylabel("Total Sales")
plt.xlabel("Shopping Mall")
plt.legend(title="Category")
plt.show()
```
